```python
# Initialiser une liste pour stocker le contenu de tous les documents chinois
#all_zh_text = []
# Initialiser une liste pour stocker les phrases de tous les documents chinois
#all_zh_sentences = []

# Parcourir tous les fichiers "Zh*.txt"
for infile in glob.glob("../Corpus/Ch*.txt"):
    with open(infile, encoding='utf-8') as f:  # S'assurer de spécifier l'encodage si nécessaire
        text = f.read()
        # Tokenization du texte avec Jieba
        tokens_zh = jieba.cut(text)
        
# Calculer la taille en nombre de tokens
words_zh = list(tokens_zh)
# Afficher le nombre de tokens
print("Nombre total de tokens dans les documents chinois :", len(words_zh))
```

```python

# Récupérer la liste des chemins des fichiers commençant par "Ch" dans le répertoire spécifié
file_paths_zh = glob.glob("../Corpus/Ch*.txt")

# Fonction pour découper le texte en phrases
def segment_texts_to_sentences_from_files(file_paths_zh):
    all_sentences = []
    # Parcourir tous les chemins de fichiers
    for file_path in file_paths_zh:
        # Ouvrir le fichier et lire le texte
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        # Découper le texte en phrases avec Jieba
        words = list(jieba.cut(text))
        sentences = []
        # Initialiser une phrase vide
        sentence = []
        # Parcourir tous les mots
        for word in words:
            # Si le mot est une ponctuation qui indique la fin d'une phrase
            if word in ['。', '？', '！']:
                # Ajouter la phrase courante à la liste des phrases
                sentences.append(''.join(sentence))
                # Réinitialiser la phrase courante
                sentence = []
            else:
                # Ajouter le mot à la phrase courante
                sentence.append(word)
        # Si la phrase finale n'est pas vide, l'ajouter à la liste des phrases
        if sentence:
            sentences.append(''.join(sentence))
        # Ajouter les phrases du document courant à la liste de toutes les phrases
        all_sentences.extend(sentences)
    return all_sentences

# Exemple d'utilisation avec les fichiers commençant par "Ch"
sentences_zh = segment_texts_to_sentences_from_files(file_paths_zh)
#print(sentences)
# Afficher le nombre de phrases
print("Nombre total de phrases dans les documents chinois:", len(sentences_zh))
```

```python
# On a trois ensembles d'entraînement pour chaque langue : train_fr, train_en, train_zh
# Et une liste de langues correspondantes : ['fr', 'en', 'zh']

# Créer des listes vides pour stocker les données d'entraînement combinées et les étiquettes de langue
combined_train_data = []
labels_train = []

# Parcourir chaque ensemble de données d'entraînement
for lang, train_set in zip(['FRENCH', 'ENGLISH', 'CHINESE_ZH'], [train_fr, train_en, train_zh]):
    # Pour chaque phrase dans l'ensemble de données d'entraînement actuel
    for sentence in train_set:
        # Ajouter la phrase à la liste principale de données d'entraînement combinées
        combined_train_data.append(sentence)
        # Ajouter l'étiquette de langue correspondante à la liste des étiquettes
        labels_train.append(lang)

# Vérifier que les données et les étiquettes sont bien combinées
print("Exemple de données d'entraînement combinées :", combined_train_data[:5])
print("Exemple d'étiquettes correspondantes :", labels_train[:5])
```

```python
# Avec CountVectorizer (BoW Bag of Words)

class LanguageDetector_BoW:
    def __init__(self, n_features=256):
        self.vectorizer = CountVectorizer(analyzer='char', max_features=n_features)
        self.models = {}
        self.n_features = n_features

    def extract_features(self, texts):
        X = self.vectorizer.transform(texts)
        X = X.multiply(1 / X.sum(axis=1))  # Normalizing each row
        return X.tocsr()  # Convert to CSR format for better row slicing

    def train_gmm_models(self, data, labels, min_components=3):
        self.vectorizer.fit(data)
        features = self.extract_features(data)
        unique_labels = set(labels)
        for label in unique_labels:
            indices = np.where(np.array(labels) == label)[0]
            model_data = features[indices, :]  # Correct slicing using CSR matrix
            n_samples = model_data.shape[0]
            n_components = max(min_components, int(np.sqrt(n_samples) / 2))  # Dynamic number of components
            if n_samples < n_components:
                print(f"Not enough samples to train GMM for {label}. Required: {n_components}, Found: {n_samples}")
                continue
            gmm = GaussianMixture(n_components=n_components, covariance_type='full', random_state=42)
            gmm.fit(model_data.toarray())  # Need to convert to dense array for GMM fitting
            self.models[label] = gmm

    def predict_language(self, text):
        features = self.extract_features([text])
        best_label = None
        max_language_score = float('-inf')
        for label, model in self.models.items():
            language_score = model.score(features.toarray())  # Convert to dense array for scoring
            if language_score > max_language_score:
                max_language_score = language_score
                best_label = label
        return best_label
```

```python
# # Usage example

detector_BoW = LanguageDetector_BoW()
detector_BoW.train_gmm_models(combined_train_data, labels_train)

random_sentence = random.choice(combined_test_data)
predicted_language = detector_BoW.predict_language(random_sentence)
print(f"PREDICTED LANGUAGE OF '{random_sentence}' IS {predicted_language}.")
```

```python
labels_predicted = []
labels_language = set(labels_test)
compare = {}

for sentence in combined_test_data:
    predicted_language = detector_BoW.predict_language(sentence)
    labels_predicted.append(predicted_language)

for i, element in enumerate(labels_test):
    if labels_predicted[i] == element:
        compare[element] = compare.get(element, 0) + 1

for label in compare:
    print("Précision {label}: {score}".format(
        label = label,
        score = compare[label]/len([x for x in labels_predicted if x == label]))
        )
    
print("")
for label in compare:
    print("Rappel {label}: {score}".format(
        label=label,
        score=compare[label]/len([x for x in labels_test if x == label]))
        )
```
