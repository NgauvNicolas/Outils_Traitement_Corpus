```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from glob import glob
import glob
import nltk
import random
import jieba
```

```python
# Initialiser une liste pour stocker le contenu de tous les documents français
all_fr_text = []
# Initialiser une liste pour stocker les phrases de tous les documents français
all_fr_sentences = []
# Initialiser une liste pour stocker les tokens de chaque document français
#tokenized_docs_fr = []

# Parcourir tous les fichiers "Fr*.txt"
for infile in glob.glob("../Corpus/Fr*.txt"):
    with open(infile, encoding='utf-8') as f:  # S'assurer de spécifier l'encodage si nécessaire
        text = f.read()
        all_fr_text.append(text)
        # Convertir le générateur de phrases en liste et l'ajouter à la liste des phrases
        all_fr_sentences.extend(nltk.sent_tokenize(text))

# Concaténer le contenu de tous les documents fr en un seul document
concatenated_text_fr = '\n'.join(all_fr_text)

# Tokenizer le texte
tokens_fr = nltk.word_tokenize(concatenated_text_fr)
# Liste de token du texte
#tokenized_docs_fr.append(tokens_fr)

# Calculer la taille en nombre de jetons
token_fr_count = len(tokens_fr)

# Afficher le nombre de tokens
print("Nombre total de tokens dans les documents fr:", token_fr_count)
#print("Liste de tous les tokens des documents fr:", tokenized_docs_fr)
#print(len(tokenized_docs_fr))
# Afficher le nombre de phrases
print("Nombre total de phrases dans les documents français:", len(all_fr_sentences))
# Afficher les premières phrases (facultatif)
print("Premières phrases dans les documents français:")
for sentence in all_fr_sentences[:5]:
    print(sentence)
```

```python
import glob

# Récupérer la liste des chemins des fichiers commençant par "Ch" dans le répertoire spécifié
file_paths_zh = glob.glob("../Corpus/Ch*.txt")

# Fonction pour découper le texte en phrases
def segment_texts_to_sentences_from_files(file_paths_zh):
    all_sentences = []
    # Parcourir tous les chemins de fichiers
    for file_path in file_paths_zh:
        # Ouvrir le fichier et lire le texte
        with open(file_path, 'r', encoding='utf-8') as file:
            text = file.read()
        # Découper le texte en phrases avec Jieba
        words = list(jieba.cut(text))
        sentences = []
        # Initialiser une phrase vide
        sentence = []
        # Parcourir tous les mots
        for word in words:
            # Si le mot est une ponctuation qui indique la fin d'une phrase
            if word in ['。', '？', '！']:
                # Ajouter la phrase courante à la liste des phrases
                sentences.append(''.join(sentence))
                # Réinitialiser la phrase courante
                sentence = []
            else:
                # Ajouter le mot à la phrase courante
                sentence.append(word)
        # Si la phrase finale n'est pas vide, l'ajouter à la liste des phrases
        if sentence:
            sentences.append(''.join(sentence))
        # Ajouter les phrases du document courant à la liste de toutes les phrases
        all_sentences.extend(sentences)
    return all_sentences

# Exemple d'utilisation avec les fichiers commençant par "Ch"
sentences_zh = segment_texts_to_sentences_from_files(file_paths_zh)
#print(sentences)
# Afficher le nombre de phrases
print("Nombre total de phrases dans les documents chinois:", len(sentences_zh))
```

```python
# Split the corpus in 3 parts : dev, train and test for the 3 languages
# 10% for dev, 80% for train, 10% for test



import random
random.seed(42)


# dev
len_dev_fr = int(len(all_fr_sentences) * 0.1)
print("Nombre total de phrases dans les documents français:", len(all_fr_sentences))
#print(len_dev_fr)
print("\n")

len_dev_en = int(len(all_en_sentences) * 0.1)
print("Nombre total de phrases dans les documents anglais:", len(all_en_sentences))
#print(len_dev_en)
print("\n")

len_dev_zh = int(len(sentences_zh) * 0.1)
print("Nombre total de phrases dans les documents chinois:", len(sentences_zh))
#print(len_dev_zh)
print("\n")
#print(len_docs)
#print(len_dev)


dev_fr = random.sample(all_fr_sentences, len_dev_fr)
remain_fr = [d for d in all_fr_sentences if d not in dev_fr]
#print(len(remain_fr))
#print(len(dev_fr))
#print("\n")

dev_en = random.sample(all_en_sentences, len_dev_en)
remain_en = [d for d in all_en_sentences if d not in dev_en]

dev_zh = random.sample(sentences_zh, len_dev_zh)
remain_zh = [d for d in sentences_zh if d not in dev_zh]
#print(len(remain))


# test
test_fr = random.sample(remain_fr, len_dev_fr)
test_en = random.sample(remain_en, len_dev_en)
test_zh = random.sample(remain_zh, len_dev_zh)
#print(len(test))



# train
train_fr = [d for d in remain_fr if d not in test_fr]
train_en = [d for d in remain_en if d not in test_en]
train_zh = [d for d in remain_zh if d not in test_zh]

print(f"Len dev_fr : {len(dev_fr)}, len test_fr : {len(test_fr)}, len train_fr : {len(train_fr)}")
print(f"Len dev_en : {len(dev_en)}, len test_en : {len(test_en)}, len train_en : {len(train_en)}")
print(f"Len dev_zh : {len(dev_zh)}, len test_zh : {len(test_zh)}, len train_zh : {len(train_zh)}")
```
